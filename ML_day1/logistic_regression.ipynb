```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

# create 200 samples that will fit into one of 2 classes. each sample has 2 features.
# 0 of these are redundant, and 2 are informative. for each of the 2 classes, there will be 1 cluster.
X, y = make_classification(n_samples=200, n_classes=2, n_features=2, 
                           n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)

```

```python
h = .02  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # get all of X (array of dimension 2), then select only the 0th index for the x co-ords
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # get all of X, then select the first index for all y co-ords
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),  # arange is like the linspace of matplot, [min, max)
                     np.arange(y_min, y_max, h))  # meshgrid taking x and y from both linspaces
plt.scatter(X[:,0],X[:,1],c=y,cmap="RdBu")  # data positions X[0] and X[1] co-ordinates. spots are red or blue based on membership of shapes in y
plt.xlim((-5,5))
plt.ylim((-5,5))
plt.show()
```

```python
clf = LogisticRegression().fit(X, y)
```

```python
w = clf.coef_[0]  # logreg.coef_ returns coefficient of (true/false)  
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
plt.xlim((-5,5))
plt.ylim((-5,5))
yy = a * xx - (clf.intercept_[0]) / w[1]  
plt.scatter(X[:,0],X[:,1],c=y,cmap="RdBu")
plt.plot(xx, yy, 'k-')
plt.show()
```
